# Importing libraries

```python
import pandas as pd
```

# Loading and Cleaning data

```python
df = pd.read_csv('data/data.csv',sep=';',encoding='latin1')
```

```python
df.columns = df.columns.str.lower()
```

```python
df.drop(columns=['fecha_corte','periodo_registro'],inplace=True)
```

```python
df.columns
```

```python
obj_cols = df.select_dtypes('object').columns
```

```python
df.loc[:,obj_cols] = df[obj_cols].apply(lambda x: x.str.lower())
```

```python
temp = {'soltero/a':'soltero',
        'casado/a':'casado',
        'divorciado/a':'divorciado',
        'viudo/a':'viudo'}

df.loc[:,'estado_civil'] = df['estado_civil'].replace(temp)
```

```python
temp = {'día (d)':'dia',
        'noche (n)':'noche',
        'tarde (t)':'tarde'}

df.loc[:,'turno_trabajo'] = df['turno_trabajo'].replace(temp)
```

```python
df['gravedad_accidente'].value_counts()
```

```python
df['fecha_accidente'] = pd.to_datetime(df['fecha_accidente'],format='%d-%m-%Y')
```

```python
df.groupby('dia').size()
```

```python
df['dia'] = df['fecha_accidente'].dt.day_name()
```
